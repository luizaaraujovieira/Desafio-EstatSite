```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('Iris.csv')
```

```python
## usando seaborn para plotar
## diferenciando as cores por grupos de espécies com o hue
sns.lmplot( x="SepalLengthCm", y="PetalLengthCm", data=df, fit_reg=False, hue='Species', legend=False)
plt.legend()
 
```
