```python
## Exercício 8

## Utilizando o dataset Iris: Primeiro, apague a substring “Iris-” 
## da coluna Species. Em seguida, adicione 3 novas colunas à tabela 
## inicial, sendo que cada coluna receberá uma dummy referente a cada
## uma das species. Ou seja, você deve criar uma coluna chamada 
## Dummy_Setosa, que recebe 1 se a flor for da espécie Setosa e 0 
## caso contrário. O mesmo para as demais espécies.
```

```python
import pandas as pd
import numpy as np
data2 = pd.read_csv('Iris.csv')
```

```python
## limpando Iris- do nome
data2['Species'] = data2['Species'].str.replace('Iris-', '')
```

```python
## pegando por espécie desejada e transformando em True/False
data2['Dummy_Setosa'] = data2['Species'].str.contains('setosa')

## multiplicando por 1. assim true vira 1 e false 0
data2['Dummy_Setosa'] = data2['Dummy_Setosa']*1
```

```python
## pegando por espécie desejada e transformando em True/False
data2['Dummy_Virginica'] = data2['Species'].str.contains('virginica')

## multiplicando por 1. assim true vira 1 e false 0
data2['Dummy_Virginica'] = data2['Dummy_Virginica']*1
```

```python
## pegando por espécie desejada e transformando em True/False
data2['Dummy_Versicolor'] = data2['Species'].str.contains('versicolor')

## multiplicando por 1. assim true vira 1 e false 0
data2['Dummy_Versicolor'] = data2['Dummy_Versicolor']*1
```
